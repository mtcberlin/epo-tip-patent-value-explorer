```python
import subprocess, os
from pathlib import Path
from IPython.display import display, HTML, clear_output

REPO_URL = "https://github.com/mtcberlin/patent-value-explorer.git"
APP_DIR = Path.home() / "patent_value_explorer"

def _status(msg):
    clear_output(wait=True)
    display(HTML(f"""
    <div style="font-family: system-ui, sans-serif; padding: 20px;
                background: #f8fafc; border-left: 4px solid #082453;
                border-radius: 8px;">
        <div style="font-size: 15px; color: #334155;">⏳ {msg}</div>
    </div>"""))

if (APP_DIR / ".git").is_dir():
    _status("Updating Patent Value Explorer...")
    subprocess.run(["git", "-C", str(APP_DIR), "pull", "--ff-only"],
                   capture_output=True, text=True)
else:
    _status("Downloading Patent Value Explorer (first run)...")
    subprocess.run(["git", "clone", REPO_URL, str(APP_DIR)],
                   capture_output=True, text=True, check=True)

os.chdir(str(APP_DIR))

launch_path = str(APP_DIR / "launch.py")
exec(compile(open(launch_path).read(), launch_path, "exec"), {**globals(), "__file__": launch_path})
```
